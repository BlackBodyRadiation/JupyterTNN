```julia
b = SpinBasis(1//2)
sx = 1/2*sigmax(b)
sy = 1/2*sigmay(b)
sz = 1/2*sigmaz(b)
XX = sx ⊗ sx 
YY = sy ⊗ sy 
ZZ = sz ⊗ sz
```

```julia
Jxy=1
Jz=1
```

```julia
function basis(N)    
    spinchain=b
    for ispin in 2:N
        spinchain = spinchain ⊗ b
        
    end
    return spinchain # same local
end
```

```julia
basis(6)    
```

```julia
function CrtModelH(N)
    Htwoint=Jxy*(XX+YY)+Jz*ZZ
    HXXZ = embed(basis(N), [1,2] , Htwoint)
    for s in 2:N-1
        HXXZ += embed(basis(N), [s,s+1] , Htwoint)
    end
    return HXXZ
end
```

```julia
CrtModelH(7)
```

```julia
function solvemodl(N)
    E5 = eigenenergies(CrtModelH(N), 5)
    println(" ")
    #println("Lowest five eigenvalues");
    #println(E5);
    println(" ")
    println(" ")
    # full spectrum
    #EE, UU = eigenstates(dense(CrtModelH(N)))
    println(" ")
    #println("All Eigenvalues");
    #println(EE);
    println(" ")
    return E5
end
```

```julia
function printmodls(N)
    n=Any[]
    for i in 3:N
        push!(n, solvemodl(i))
    end
    return n
end
        
```

```julia
function Phi0(N) 
    sup = spinup(b)
    sdo = spindown(b)
    psi0 = sup
    for ispin in 2:N
        if ispin == 5
           psi0 = psi0 ⊗ sdo
        else 
           psi0 = psi0 ⊗ sup
        end 
    end

    return psi0
end
```

```julia
Nlist = 7:10;
times = 0:.1:6
psit=[]; 
toutt=[]; 
for l in Nlist
    tout, psi_t = timeevolution.schroedinger(times, Phi0(l) , CrtModelH(l))
    push!(psit,psi_t)
    push!(toutt,tout)
end 
```
